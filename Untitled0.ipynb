```python
df['median_house_value'].min(), df['median_house_value'].max()
```

```python
min = df['median_house_value'].min()
df[df['median_house_value'] == min]['population'].max()   # первый вариант решения
```

```python
df[df['median_house_value'] == df['median_house_value'].min()]['population'].max()   # второй вариант решения
```

```python
max(df[df['median_house_value'] == df['median_house_value'].min()]['population']) # нестабильный вариант решения, из-за строк
```
